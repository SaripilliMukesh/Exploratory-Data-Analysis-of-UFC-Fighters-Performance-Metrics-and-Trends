# **Feature engineering**

Enhancing the UFC fighters dataset with additional performance metrics

```python
df["matches_played"] = df["wins"] + df["losses"] + df["draws"]
df["win_percentage"] = round((df["wins"] / df["matches_played"]),2) * 100
df["los_percentage"] = round((df["losses"] / df["matches_played"]),2) * 100
df.head()
```

**Assigning weight classes to UFC fighters based on their weight in kilograms**

Flyweight                  125 lb          57 kg

Bantamweight         135 lb           61 kg

Featherweight          145 lb          66 kg

Lightweight              155 lb           70 kg

Welterweight            170 lb           77 kg

Middleweight           185 lb           84 kg

Light Heavyweight   205 lb           93 kg

Heavyweight            205-265 lb    93-120 kg

```python
df['weight_in_kg']=df['weight_in_kg'].round(1)
```

```python
weight_classes = {
'Flyweight':56.7,
'Bantamweight':	61.2,
'Featherweight': 65.8,
'Lightweight':70.3,
'Welterweight':77.1,
'Middleweight':	83.9,
'Light heavyweight':93.0,
'Heavyweight':120.2}


def assign_weight_class(weight):
    for weight_class,uperlimit in weight_classes.items():
        if weight <= uperlimit:
            return weight_class

df['weight_class'] = df['weight_in_kg'].apply(assign_weight_class)
df.head()
```

**BMI Calculation:**


Description: Calculate the Body Mass Index (BMI) using the Height and Weight features. BMI can be a significant indicator of a fighter's physical condition and readiness.

Formula: BMI = Weight (kg) / (Height (m))^2.

Purpose: BMI can provide insights into how fighters' physical fitness relates to their success in the octagon

```python
df['BMI'] = df['weight_in_kg'] / (df['height_cm'] / 100) ** 2
```

# **Correlation Analysis**

**Visualizing correlations between physical attributes, performance metrics, and fight outcomes among UFC fighters**

**Physical Attributes**
Age,
Height (cm),
Weight (kg)


**Performance Metrics**
Significant Strikes Landed per Minute,
Average Takedowns Landed per 15 Minutes,
Average Submissions Attempted per 15 Minutes

```python
physical_attributes = ['age','height_cm', 'weight_in_kg']
performance_metrics = ['significant_strikes_landed_per_minute','average_takedowns_landed_per_15_minutes','average_submissions_attempted_per_15_minutes']
outcome = ['wins', 'losses', 'draws']

def physical_attributes_corr_plot(df):
  corr_matrix= df[physical_attributes+outcome].corr()
  sns.heatmap(corr_matrix,annot=True)
  return plt.show()

def performance_metrics_corr_plot(df):
  corr_matrix= df[performance_metrics+outcome].corr()
  sns.heatmap(corr_matrix,annot=True)
  return plt.show()

physical_attributes_corr_plot(df)
performance_metrics_corr_plot(df)
```

**Relationship between fighter weight and wins across different UFC weight classes**

```python
# df[df['weight_class']=='Lightweight'][['weight_in_kg','wins']].plot.scatter(x='weight_in_kg',y='wins')
# print(df[['weight_in_kg','wins']].corr())
def weight_weight_class_scatter(df):
  weight_classes=df['weight_class'].unique()
  for weight_class in weight_classes:
    df[df['weight_class']==weight_class][['weight_in_kg','wins']].plot.scatter(x='weight_in_kg',y='wins')
    plt.axvline(df[df['weight_class']==weight_class]['weight_in_kg'].mean(), color='red')
    plt.title(weight_class)
    plt.xlabel('Weight (kg)')
    plt.ylabel('Wins')
    plt.show()
    print(df[df['weight_class']==weight_class][['weight_in_kg','wins']].corr())
    print((df[df['weight_class']==weight_class]['weight_in_kg'].mean().round(1)))
weight_weight_class_scatter(df)
```

# **Web Scraping**

```python
import requests
from bs4 import BeautifulSoup

url = "http://statleaders.ufc.com/en/career"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

article = soup.find('article', class_='results-group Time-group',id="ControlTime-group")

ranks = []
fighters = []
times = []


rows = article.find_all('div', class_='results-table--tr')[1:]
for row in rows:
    rank = row.find_all('span')[0].text
    fighter = row.find_all('span')[1].text
    time = row.find_all('span')[2].text

    ranks.append(rank)
    fighters.append(fighter)
    times.append(time)

control_time_df = pd.DataFrame({'Rank': ranks,'Fighter': fighters,'Time': times})
control_time_df
```

```python
url = 'https://www.ufc.com/rankings'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

title_elements = soup.find_all(class_='views-field views-field-title')
entries = []
for element in title_elements:
    entry = element.get_text().strip()
    entries.append(entry)

pound4pound_df = pd.DataFrame({'Fighter': entries})
pound4pound_df=pound4pound_df.head(15)
pound4pound_df
```
